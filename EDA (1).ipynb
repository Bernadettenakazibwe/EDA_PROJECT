```python
 #Calculate Z-score for each value in 'heartRate'
df['z_score'] = (df['heartRate'] - df['heartRate'].mean()) / df['heartRate'].std()

# Find outliers
outliers = df[(df['z_score'] > 3) | (df['z_score'] < -3)]
print("Outliers detected using Z-Score:")
print(outliers)
```

```python
 #Calculate Z-score for each value in 'ppgGreen'
df['z_score'] = (df['ppgGreen'] - df['ppgGreen'].mean()) / df['ppgGreen'].std()

# Find outliers
outliers = df[(df['z_score'] > 3) | (df['z_score'] < -3)]
print("Outliers detected using Z-Score:")
print(outliers.sum())
```

```python
#Removing outliers using the Z-Score 

def remove_outliers_zscore(df, threshold=3):
    numeric_cols = df.select_dtypes(include=[np.number]).columns  # Select numerical columns
    df_clean = df.copy()  # Make a copy of the original DataFrame
    
    for col in numeric_cols:
        # Calculate Z-scores for the column
        z_scores = (df_clean[col] - df_clean[col].mean()) / df_clean[col].std()
        
        # Keep only rows where Z-score is within the threshold
        df_clean = df_clean[(z_scores < threshold) & (z_scores > -threshold)]
    
    return df_clean

# Apply function to remove outliers
df_cleaned = remove_outliers_zscore(df)

# Show results
print("Original DataFrame:")
print(df)
print("\nDataFrame after removing outliers:")
print(df_cleaned)
```

```python
#Viewing if most of the outliers are removed 
columns= ['heartRate','ppgGreen']
for col in columns:
    plt.boxplot(df_cleaned[col])
    plt.title(f'Boxplot for {col}')
    plt.ylabel('Heart Rate')
    plt.show()
 
```
